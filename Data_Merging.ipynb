```python
import numpy as np
import pandas as pd
from scipy.io import arff
import os
```

```python
filenames=os.listdir('data2005')
print(filenames)
```

```python
data=arff.loadarff('data2005/entry01.weka.allclass.arff')
df=pd.DataFrame(data[0])

df.head()
```

```python
data[1]
```

```python
dfs=[]
for file in filenames:
    data=arff.loadarff('data2005/'+ file)
    dfs.append(pd.DataFrame(data[0]))
df=pd.concat(dfs,ignore_index=True)
```

```python
df.to_csv('data_merged.csv',index=False)
```
